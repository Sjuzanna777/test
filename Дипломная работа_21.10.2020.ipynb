2. Гипотеза: "Чем больше количество субъектов МСП было сосредоточено в регионе, тем ниже уровень безработицы населения"

```python
Data_finally_2 = Data[Data['Year']==2018].iloc[:, 2:19]
```

```python
unemployment = Data_finally_2['Уровень безработицы (по методологии МОТ) 15-72 лет, %']
Count_sme = Data_finally_2['Реестр МСП, на конец отчетного года']
sns.regplot(x= Count_sme , y = unemployment, data = Data_finally , fit_reg=True)
plt.xlabel('Количество субъектов МСП')
plt.ylabel('Уровень безработицы (по методологии МОТ) 15-72 лет, %')
plt.title('Рис.2. Корреляционная зависимость двух выборок')
plt.grid(True)
plt.show()
```

Корреляция между уровнем безработицы и количеством субъектов в регионе отрицательная. Можно сформулировать следующий вывод: чем большее количество субъектов МСП, тем ниже уровень безработицы в регионе и выше уровень ВРП на душу населения. Благоприятные условия для введения деятельности субъектов МСП в том или ином регионе оказывают позитивное влияние на экономику региона - места деятельности.

```python
VRP = Data_finally['ВРП  на душу населения, руб.']
investment = Data_finally['Инвестиции в ОК на душу населения , руб.']
```

```python
VRP = np.array(VRP).reshape(-1,1)
investment = np.array(investment).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Разделить данные на 80% тренировочных, 20% тестовых

```python
from sklearn.model_selection import train_test_split

x = VRP,
y = investment
X_train, X_test, y_train, y_test = train_test_split(
    VRP,
    investment,
    test_size = 0.2
)
```

```python
slr = LinearRegression()

slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

print('MSE train: {:.3f}, test: {:.3f}'.format(
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

```python
plt.scatter(y_train_pred,  y_train_pred - y_train,
            c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred,  y_test_pred - y_test,
            c='lightgreen', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.xlim([0, 500000])
plt.ylim([0, 50000])
plt.tight_layout()
```
